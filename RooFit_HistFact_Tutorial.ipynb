## Setting up model

RoorRealVar is a constructor that has many optional argruments
(const char * name, const char * title, double value, double minValue, double maxValue, const char * unit = "" )
So x can contain real numbers between -10 and 10 while mean is also a real number between -10 and 10 but also has an assigned current value of 1

A pdf is the probability density function and in this simple case will simply be a gaussian.

A frame is TH1 object and is used to configure plot randes, configure axes, etc. It can hold any number of objects such as histograms or fit model curves. It is created by attaching a frame to an observable variable such as invarient mass.

```python
# Gauss parameters
x = ROOT.RooRealVar("x", "x", -10, 10)
mean  = ROOT.RooRealVar("mean",  "mean",  5279, 5200, 5400)  # B⁰ mass in MeV
sigma = ROOT.RooRealVar("sigma", "sigma", 20,   5,    100)

# PDF of a Gaussian
gauss = ROOT.RooGaussian("gauss", "gaussian PDF", x, mean, sigma)

# Attach frame to the x variable. Practically x would represent something in phase space.
xframe = x.frame(Title = "Gaussian PDF")
```

## Plotting basics

First attach the desired graphic objects to the frame and draw that frame

Then create a canvas and draw the that canvas. 

Once an object has been drawn to a frame, it stays on that frame. So drawing a second object to the same frame and drawing that frame on a second canvas will draw both objects.

```python
gauss.plotOn(xframe)
xframe.Draw()
```

```python
c = ROOT.TCanvas("test")
xframe.Draw() 

c.Draw()

sigma.setVal(3)                      # Manage dataset
gauss.plotOn(xframe, LineColor='r')  # plot dataset on frame
cc = ROOT.TCanvas("ttest")           # make canvas
xframe.Draw()                        # draw frame
cc.Draw()                            # draw canvas
```

```python
c.Close()
cc.Close()
```

## Fitting data

First generate a RooDataSet.

Then 

```python
# Create 10000 events of x ie between -10, 10.
data = gauss.generate({x}, 10000)  # ROOT.RooDataSet
data.Print("V")
for i in range(10):
    data.get(i).Print("V")
```

```python
xframe2 = x.frame(Title="Gaussian pdf with data")  # RooPlot
data.plotOn(xframe2)
gauss.plotOn(xframe2)
c = ROOT.TCanvas("test")
xframe2.Draw()

c.Draw()
```

```python
c.Close()
```

```python
gauss.fitTo(data, PrintLevel=-1)
```
